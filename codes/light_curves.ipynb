```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# signal = np.array([1, 10, 1, 10, 1, 10, 1, 10, 1, 10], dtype=float)
signal=np.array([10,1,1,1,1]*20, dtype=float)
fourier = np.fft.rfft(signal)
n = signal.size

freq = np.fft.rfftfreq(n)
print(freq)
```

```python

N = 256
t = np.arange(N)

m = 4
nu = float(m)/N
f = np.sin(2*np.pi*nu*t)

f[f<0.2]=0.2
plt.plot(t,f)
plt.show()
ft = np.fft.fft(f)
freq = np.fft.fftfreq(N)
plt.plot(freq, ft.real**2 + ft.imag**2)
plt.show()

```
